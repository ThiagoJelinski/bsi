```python
## Importe a biblioteca matplotlib e crie um gráfico de barras simples com alguns dados fictícios.

import matplotlib.pyplot as plt

foods = ["Baconzitos", "Nescau", "Nutella", 'Subway']
calories = [500, 400, 450, 600]
bar_colors = ['tab:red', 'tab:orange', 'tab:brown', 'tab:blue']

fig, ax = plt.subplots()

bars = ax.barh(foods, calories, color=bar_colors)

ax.set_xlabel('Calories (per 100g)')
ax.set_title("Caloric Analysis of Processed Foods (per 100g)")

for bar, cal in zip(bars, calories):
    ax.text(bar.get_width(), bar.get_y() + bar.get_height() / 2, f'{cal} cal', va='center')

legend_labels = ['Baconzitos (red)', 'Nescau (orange)', 'Nutella (brown)', 'Subway (blue))']
ax.legend(bars, legend_labels, loc='upper right', title="Food Color")

plt.show()
```

```python
## Escreva um programa que importe o pacote matplotlib e plote a função seno.

import matplotlib.pyplot as plt
from math import radians
import numpy as np

x = np.arange(radians(-360), radians(360), 0.1)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y,x,z)

plt.title("Gráfico de Seno e Cosseno")
plt.legend(["Seno de x", "Cosseno de x"])
plt.xlabel("De -360º para 360º")
plt.ylabel("sin(x) e cos(x)")

plt.show()
```

```python
## Escreva um programa que importe o pacote beautifulsoup4 e faça a raspagem de dados (web crawling) de uma página HTML.

import requests
from bs4 import BeautifulSoup

url = "https://www.hackerrank.com/domains/sql"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36"
}
response = requests.get(url, headers=headers)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, "html.parser")
    challenges_list = soup.find("div", class_="challenges-list")

    if challenges_list:
        challenge_items = challenges_list.find_all("a", class_="challenge-list-item")
        num_sql_challenges = len(challenge_items)

        print(f"No total há {num_sql_challenges} de desafios de SQL")
    else:
        print("Não foi possível encontrar a lista de desafios.")
else:
    print("Falha na solicitação HTTP.")
```
